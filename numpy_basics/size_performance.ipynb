```python
# We can go ahead and look at the performance.

# When we perform calculations which ones are executed faster
```

```python
# We create an array of random integers

a = np.arange(1000)

# We create a list of 1000 numbers

l = list(range(1000))
```

```python
# We try and perfrm the same opeartion on the list and the array and see which one will be executed faster


# sum of the square of the array or list

%time np.sum( a ** 2)
```

```python
# We do the same for lists

%time sum([ x ** 2 for x in l])
```
