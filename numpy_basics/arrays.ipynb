```python
# In this notebook we will be looking at:
# Creating one dimensional arrays(HAVING ONLY ONE ROW)
# The basic data-types of arrays in numpy
# Indexing ans slicing of one dimensional arrays(HAVE ONLY ONE ROW)
```

```python
import numpy as np
```

```python
# We create an array of integers
# Arrays should have the same data-type

integers_array = np.array([1 , 2,3,4,5,6,7])
integers_array
```

```python
# We check the datatype of the array we just cdreated. Should be int64

integers_array.dtype
```

```python
# Create a another array of floats

floats_array = np.array([.5 , .4 ,.6 , .7])
floats_array
```

```python
# Check the dataype of the array we created should be float64
 
floats_array.dtype
```

```python
# When creating an array despite them bing integers you may want to change the datatype of the array

change_dtype_array = np.array([3,4,5,6,7,8] , dtype= 'float64')
change_dtype_array
```

```python
# We may want to change the data-type again
change_dtype_array.astype('int64')
```

```python
# We create another array of strings

strings_array = np.array(['A' , 'B' , 'C' , 'D'])
strings_array
```

```python
# What about a dictionary

dictionary_array = np.array({'A' : 'Pass'})
dictionary_array
```

```python
print(f"Array1 = {floats_array.dtype} \nArray2 ={integers_array.dtype} \nArray3 = {strings_array.dtype} \nArray4 = {dictionary_array.dtype}")
```

```python
# Also we can slice and index arrays just like in python

# Selectts the 1st element in the array
integers_array[0]
```

```python
# Selecting the 1st element in the floats

floats_array[0]
```

```python
# We can also slice

floats_array[0 : 2]
```

```python
# numpy also supports multi nindexing
# Like we can selcet multiple indexes at once

# Lets say we select elements at the 1st , 3rd and last index

floats_array[[ 0 , 2 , -1]]
```
