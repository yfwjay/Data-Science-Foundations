```python
# In this notebook we will be looking at :
# multideminsional arrays(having more than one row)
# The opeartions we use to manipulate multidimesional arrays
# Indexing and slicing multi-dimensional array
```

```python
import numpy as np
```

```python
# We create an array with 2 rows and several columns

multi_array = np.array([
    [1 , 2 , 3] , 
    [6 ,7 ,8] ,
])
multi_array
```

```python
# we can check the:
# shape - number of rows and columns
# size - number of elemnents in our array

print(f"Shape = {multi_array.shape} \nSize = {multi_array.size} \nData_Type = {multi_array.dtype} ")
```

```python
# We can also index and slice multi dimessiona arrays

# syntax is rows 1st then columns

```

```python
# eg

# select 1st row only. SHOULD DISPLAY 1 , 2 ,3
multi_array[0]
```

```python
# Select the second row . SHOULD DISPLAY 6 , 7 ,8

multi_array[1]
```

```python
# Now what if I wanted to select both the row and the columns

# Select the 1st row and 1st column

multi_array[0 , 0]
```

```python
# 1st row and elements from 1 upto last index

multi_array[0 , 1:]
```
