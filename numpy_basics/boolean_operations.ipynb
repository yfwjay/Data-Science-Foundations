```python
# Apart from the normal mathematical operations that we have other opearations we can perform THE BOOLEAN OPERATIONS.

# BOOLEAN OPERATIONS = > , < , >= 
```

```python
import numpy as np
```

```python
# We create an array 

b = np.arange(5)
b
```

```python
# Lets say we want to select items in the 1st index and last index

# We use multi-indexing what we learnt in the last lesson

b[[0 , -1]]
```

```python
# We can also select the indexing using boolean expressions

# The indexes we want to select we pass true the ones we want to drop we pass false

b[[True , False , False , False , True]]
```

```python
# This returns True for those which are greater than 2 and False for those less than 2
b > 2
```

```python
# If we dont want it to return true or false but the values rather we filter further

# This means for array b , select values that are greater than 2
b[b > 2]
```

```python
# mean of every value in the array

b.mean()
```

```python
# We display the values which are greater than the mean
b[b > b.mean() ]
```

```python
# We can also combine more than one boolean expression

# we display values which are equal to 0 or equal to two

# we know we should expect to see 0 and 2

c = np.array([0 , 1, 2 ,3 ,4])

c[(c == 0) | (c == 2)]

```
