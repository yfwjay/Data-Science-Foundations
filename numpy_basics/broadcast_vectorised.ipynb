```python
import numpy as np
```

```python
# We generate an array with upto 4 elemements
a = np.arange(4)
a
```

```python
# We can add each value in the array by a certain value

a + 10
```

```python
# We can perform other operations such as multipilication
a * 2
```

```python
a += 100
a
```

```python
# We can also use a loop to perform this operations

b = np.array([1 ,2, 3, 4])
b =[i + 4 for i in b]
b
```

```python
# We can also perform multiple operations on multiple arrays together depending on what we want to achieve

# We can add , multiply , divide an array with each ONLY IF THEY ARE OF THE SAME SIZE

array_1 = np.array([1 , 2 , 3])
array_2 = np.array([4 , 5 , 6 ])

array_1 + array_2

```
