```python
import numpy as np
```

```python
# RANDOM
# 1st argument is the range like any number from 0 to 10
# 2nd parameter size is the number of elements
np.random.randint( 10 , size= 20)
```

```python
np.random.rand(2  , 4)
```

```python
np.random.normal(size = 2)
```

```python
# RESHAPE

np.arange(10).reshape(2 , 5)
```

```python
# linspace

np.linspace(0 , 1 , 5)
```

```python
# Zeros

np.zeros(5)
```

```python
np.zeros((3 , 3))
```

```python
# ones

np.ones(4)
```

```python
np.ones((3 , 3))
```

```python
# Identity = equal rows and columns. 
np.identity(3)
```

```python
# eye

# Returns a 2 dimensional array with ones on the diagonal and zeros elsewhere
np.eye(3 , 3)
```

```python
np.eye(8 , 4 ,  k= 1)
# k :
# Index of the diagonal: 0 (the default) refers to the main diagonal, a positive value refers to an upper diagonal, and a negative value to a lower diagonal.
```

```python
np.eye(8 , 4 , k = -1)
```
