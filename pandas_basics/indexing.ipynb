```python
# We can create a dataframe

import pandas as pd

countries ={
    "Country" : ['Italy' , 'France' , 'Germany' , 'Kenya'] , 
    "People" : [1000 , 2000 , 3000 , 4000 ]
     }
df = pd.DataFrame(countries)
df
```

```python
# We can access specific series(column)

# Displays the 1st column and its content

df['Country']
```

```python
# We can go ahead and select specific value in that column

# lets say we want to select italy

df['Country'][0]

# we access the specific value at index 0
```

```python
# # indexing in pandas

# # Indexing in python uses iloc and loc


# Both loc and iloc are row-first, column-second. This is the opposite of what we do in native Python, which is column-first, row-second.
```

```python
# To select a row in pandas

# We will be using the dataframe i created for countries and its people

# We select the 1st row

df.iloc[0]
```

```python
# To retrieve a column we do the following using iloc

# This means we select all the rows(:) and the 1st column(0)
df.iloc[: , 0]
```

```python
# : on its own means we select everything.
#  It can be used in combination with numerical values to mean otherwise

# For example: To select upto the seconf row we could do the following:

df.iloc[:2]
```

```python
# Also we can select from the last column

# Lets read a csv file and see it in action

social_media = pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv")
social_media.head()
```

```python
# Okay to display the last column("engagement_rate") we could prceed as follows:

social_media.iloc[:3 , -1]
```

```python
# # Its much simpler to use iloc than loc

# # iloc treats our datasets like am big matrix(list of lists)

# loc, by contrast, uses the information in the indices to do its work. 
```

```python
social_media.loc[:3 , 'platform']
```
