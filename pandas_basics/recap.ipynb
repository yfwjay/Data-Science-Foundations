```python
# Group by function

# group by says put argentina countries together , Armenia together etc

grouping_data = df.groupby(['country'])['country'].count()
grouping_data
```

```python
# MAPPING
# Mapping is changing each value in a dataset and it can be achieved by using 2 functions: 
# apply() AND map()

# The difference is in how they are used.

```

```python
# map() function is for changing values of a single column

# Lets create a single column with a few valuea and try to change them using map function

values_series = pd.Series([1 ,2 ,3 ,4 ,5 ,6] , )
values_series
```

```python
# We use the map function and increase each value by 2

value_add = values_series.map(lambda x : x + 2)
value_add
```

```python
# We can create a series for names and maybe change to uppercase or lowercase etc

country_series = pd.Series(['James' , 'Joy' , 'Joan' , 'John'])
country_series
```

```python
# We change the names to upper case

country_upper = country_series.map(lambda x : x.upper())
country_upper
```

```python
# Apply function can be used both for series or dataframes

# We can apply it both on a ssingle column or multiple columns

# Create a data fram for marks

marks = {
    "English" : [50 , 70 , 60 , 90 , 40] ,
    "Maths" : [60 , 80, 90 , 30 , 50] , 
}

marks_df = pd.DataFrame(marks)
marks_df
```

```python
# We go ahead and create another column for average and try using the apply function

marks_df['Average'] = marks_df.apply(lambda row : (row['English'] + row['Maths']) / 2 , axis=1)

marks_df
```

```python
# We can go ahead and grade like above 60 thats a pass else fail

marks_df['Grade'] = marks_df.apply(lambda row : 'Yes' if row ['Average'] >= 60 else 'No' , axis= 1)
marks_df
```
