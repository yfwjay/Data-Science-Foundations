```python
# # When dealing with group_by()method one csn groupby using one column or multiple columns as we have been doing in the grouping_data.ipynb notebook


# # But when we group_by() method with multiple columns it results to = MULTIPLE_INDEXING


# # MULTIPLE_INDEXING

# # When a single row is identified by more than one column eg:

# # countries_reviewed = reviews.groupby(
# #     ['country', 'province']
# # ).description.agg([len])

# You group the data by:

# country

# province

# You count (len) the number of descriptions per group

# Pandas creates a Multi-Index using:

# Level 1 → country

# Level 2 → province
```

```python
# SEE CODE BELOW
import pandas as pd
df = pd.read_csv(r"C:\Users\HP\Desktop\Data-Science-Foundations\datasets\winemag-data_first150k.csv" , index_col=0)
df.head()
```

```python
grouped_df = df.groupby(['country' , 'province']).description.agg([len])
grouped_df
```

```python
# Why This Can Be Confusing (“Gotcha”)

# With a Multi-Index:

# You can’t access rows using just one label

# You often need two levels of indexing

# Some operations feel more complex than usual

# Example of accessing data:

try_access = grouped_df.loc['Uruguay'  , 'San Jose']
try_access
```
