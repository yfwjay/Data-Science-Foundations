```python
import pandas as pd
data = {
    "Male" : ["John" , "James" , "Victor" ], 
    "Female" : ["Joy" , "Joan" , "Joyce" ]
    
}

# Assign my data to a dataframe

df = pd.DataFrame(data)

df
```

```python
# Looking at our data frame above we may want to add an extra column loke avg_marks

df['Avg_marks'] = [80 , 50 , 30]
df
```

```python
# We can add an extra column "Grade" to know if they passed or not

df['Grade'] = ["A" , "B" , "C"]
df
```

```python
# We can create another column passed and fill it by default with no

df['Passed'] = 'No'
df
```

```python
# We go ahead and do the filtering using conditional statements

df.loc[df["Avg_marks"] >= 50 , 'Passed'] = 'Yes'
df

# df.loc[(df.Avg_marks > 50) & (df.Passed)] = 'Yes'
# df
```
