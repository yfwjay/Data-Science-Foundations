```python
# read_csv() has 30 different parameters that you can specify

# when you happen to hover on read_csv() you will see the different parameters present

# example: file_path , delimeter , index_col etc . All have different functions

```

```python
# SEP PARAMETER
# sep parameter = indicates the seperater used between the records

pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv" , sep = ",")
```

```python
# HEADER PARAMETER

# The header parameter used to specify which row do you want us to use as the column names.

# By default if you dont specify it uses the 1st row

pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv" , header = 0)
```

```python
# NAMES PARAMETER

# The paramter is used to rename the column names. You just specify the names of the columns in a list

# We assignef head = 0 so that we can alter the 1st row.

# We can also make the head = none so that we just add another row above.

pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv" , names = ['Column_1' , 'Column_2' , 'Column_3' , 'Column_4' , 'Column_5' , 'Column_6' , 'Column_7' , 'Column_8'] , header = 0)
```

```python
# INDEX_COL PARAMETER

# The parameter is used to indiacte which column you want to be used as an index that one that automatically increments

# ive indicated that we use the column post_type as the index

pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv" , index_col = "post_type")
```

```python
# USE_COLS PARAMETERS

# Only loads the parameters you have specified

# This loads only 2 columns that you happened to have specified

two_cols = pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv" , usecols = ['platform' , 'post_type'])
two_cols
```

```python
# DTYPE PARAMETER

# Defines the parameter of the a specific column

pd.read_csv(r'C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv' , dtype = {'platform' : str , 'post_type' : str})
```
