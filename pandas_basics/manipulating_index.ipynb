```python
# When reading a dataset or when you display your dataframes by default pandas creates a index(0 , 1 , 2 ,3 etc)

# But one can set their own index using set_index method and its powerful

```

```python
import pandas as pd

countries ={
    "Country" : ['Italy' , 'France' , 'Germany' , 'Kenya'] , 
    "People" : [1000 , 2000 , 3000 , 4000 ] ,
    "Description" : ["Italian_languange" , 'French_Languange' , 'German_languange' , 'Swahili_languange']
     }
df = pd.DataFrame(countries)
df
```

```python
# As you can see by default pandas creates an index

# We can set our own index based on what we wan to achieve

# by looking at our dataframe obove we can see we have a column description.

# Description is a unique column as each country has sth unique. We can use descripttion as our index . When we search for a certain description we know which country we will be talking about.


df.set_index("Description")
```
