```python
import pandas as pd

reviews = pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\winemag-data_first150k.csv" , index_col = 0)
reviews
```

```python
# There are several functions that we can use to summarize columns

# Look at column designation they have different names for each row we can use UNIQUE FUNCTION


reviews.designation.unique()
```

```python
# DESCRIBE FUNCTION

# It works both for columns with numeric values and string columns

# See below:

reviews.price.describe()
```

```python
# DESCRIBE FUNCTION IN A NON-NUMERIC COLUMN

reviews.province.describe()
```

```python
# MEAN FUNCTION

# We csn also perform a mean for s numeric column eg Points

reviews.points.mean()
```

```python
# RATIO
points_per_price_points = reviews['points'] / reviews['price']
points_per_price_points
```

```python
best_bargain_index = points_per_price_points.idxmax()
best_bargain_index
```

```python
descriptions_lower = reviews['description'].str.lower()

# Count the number of descriptions containing "tropical"
n_trop = descriptions_lower.str.contains('tropical').sum()

# Count the number of descriptions containing "fruity"
n_fruity = descriptions_lower.str.contains('fruity').sum()

# Create the final Series
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])

print(descriptor_counts)
```
