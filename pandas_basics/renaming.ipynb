```python
# In this notebook we will be dealing with renaming indexes and columns

# One can rename an index or a column depending on how you specify in the replace property

import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\HP\Desktop\Data-Science-Foundations\datasets\winemag-data-130k-v2.csv" , index_col = 0)
df.head(2)
```

```python
# We may want to rename the column points to score

df.rename(columns={'points' : 'score'}).head(1)
```

```python
# We may also want to rename the indexes

# Lets say we rename 0 to first_entry

df.rename(index={0 : 'first_entry'}).head(1)
```

```python
# One may want to give a common name to your columns and rows

#  Eg the columns you may name them as field_names
#  eg the rows you may name them as wines

# We use the rename_axis method to perform this


df.rename_axis('wines' , axis= 'rows').rename_axis('field_names' , axis= 'columns').head(1)
```
