```python
import pandas as pd

social_media = pd.read_csv(r"C:\Users\Kxbuga\Desktop\Data-Science-Foundations\datasets\social_media_engagement_dataset.csv")
social_media
```

```python
# Conditional selection is when we select certain elements based on certain conditions.

# We can select all post_length which are greater than 100

social_media.loc[social_media.post_length > 100]
```

```python
# We can combine 2 conditional statements by using an & sign

#  We display platforms whose post_length > 100 and post_type is video

social_media.loc[(social_media.post_length > 100) & (social_media.post_type == 'Video')]
```
