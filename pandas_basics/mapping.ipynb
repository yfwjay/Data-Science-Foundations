```python
# # MAPPING

# # Mapping is taking each single value in a dataset and transforming it in a consisitent way.

# eg:
# convert names to uppercase
# change currency from usd to ksh
# categoriztion and many more
```

```python
# For this to be possible there are 2 ways we can achieve this.

# By using the map() function or apply() function
```

```python
# MAP() FUNCTION

# map() appliies a function to each value in a single column(series)
# THAT IS: FOR EVERY VALUE HERE DO THIS ONE THING. WORKS FOR ONLY ONE COLUMN AND BEST FOR SIMPLE TRANSFORMATIONS

# See example below

import pandas as pd

# We create a single column with numbers 2 , ,4 ,5 ,6
column_1 =pd.Series([2 , 4 , 5 ,6])
column_1

# We may decide we increase each value in that coumn by 1


add_one = column_1.map(lambda x : x + 1)
add_one

# If you look at our column you will see that we have incremented each value in the column by 1
```

```python
# the map is best used for:
# simple calculations
# convert categories
# apply quick maths
# replace values using a dictionary

country = pd.Series(['USA' , 'Germany' , 'France'] , name = 'Countries')
country
```

```python
# we replace the columns for countries with different countries using a dictionary

new_countries = country.map({
    "USA" : "Kenya" ,
    "Germany" : 'Uganda' , 
    "France" : 'Tanzania'
})
new_countries
```

```python
# lets see if we can use a lambda to cjhange to lower case

country_lambda = country.map(lambda x : x.lower())
country_lambda
```

```python
# APPLY FUNCTION

# This is a flexible mapping(Series or DataFrame)

# apply() applies to a function
#  to each value
#  each row
#  each column

#  DEPENDS ON HOW YOU USE IT

# IT WORKS IN SERIES OR DATAFRAMES
```

```python
# APPLY ON SERIES

# We use the series column_1

column_apply = column_1.apply(lambda x : x ** 2)
column_apply

# We square each value in our column
```

```python
# APPLY ON DATAFRAMES

df = pd.DataFrame({
    'English' : [60 ,70 , 80 ,90 ] , 
    'Math' : [90 ,50 ,70 ,40]
})
df
```

```python
# We see Apply in action

# Perform average of english and math

df['Average'] = df.apply(lambda row: (row['English'] + row['Math']) / 2 , axis = 1)

# axis = 1 means apply per row
df
```

```python


df['Passed'] = df.apply(lambda row : 'Yes' if row['Average'] >70 else 'No' , axis = 1)
df
```
